```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# download required datasets (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

print("Chatbot: Hello! I am an NLP chatbot. Type 'bye' to exit.")

while True:
    
    user_input = input("You: ").lower()
    
    # Tokenization
    tokens = word_tokenize(user_input)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in tokens if word not in stop_words]

    # Rule-based responses using if-else
    if "hello" in tokens or "hi" in tokens:
        print("Chatbot: Hello! How can I help you?")

    elif "name" in tokens:
        print("Chatbot: My name is NLP Chatbot.")

    elif "help" in tokens:
        print("Chatbot: I can answer questions about greetings, my name, or basic help.")

    elif "weather" in tokens:
        print("Chatbot: I cannot check live weather yet, but it's a good day to learn Python!")

    elif "bye" in tokens:
        print("Chatbot: Goodbye! Have a great day.")
        break

    else:
        print("Chatbot: Sorry, I didn't understand that.")
```
